```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Training\\Phase 1\\Dataset\\Housing.csv")
df.head()
```

```python
correlation = df['price'].corr(df['area'], method = 'pearson' )
print(f"Pearson Correlation between Price and Area: {correlation:.4f}")
```

```python
sns.scatterplot(x=df['price'], y=df['area'])
plt.title('Scatter plot between Price and Area')
plt.xlabel("Price")
plt.ylabel("Area")
```
