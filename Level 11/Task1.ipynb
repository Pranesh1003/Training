```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
```

```python
df = pd.read_csv("C:\\Training\\Phase 1\\Dataset\\housing.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[['median_income','median_house_value']].dropna()
```

```python
x = df[['median_income']]
y = df['median_house_value']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
X_test_sorted, y_test_sorted = zip(*sorted(zip(x_test.values.flatten(), y_pred)))
plt.scatter(x_test,y_test,color="blue",alpha=0.3,s=10,label="Actual Data")
plt.plot(X_test_sorted,y_test_sorted,color='black', linestyle='--',linewidth=2,label="Regressionline")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.title("Linear Regression")
plt.legend(loc="upper right")
```

```python
r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
print(f"R2 Score: {r2:.2f}")
print(f"MeanSquaredValue: {mse:.4f}")
```

```python
if r2 > 0.7:
    print("The mode has high accuracy")
elif r2 > 0.4:
    print("The model has average accuracy")
else :
    print("The model has low accuracy")
```
